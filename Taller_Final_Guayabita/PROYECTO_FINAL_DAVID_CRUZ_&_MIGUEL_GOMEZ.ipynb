### Función de lanzamiento 

Esta función se creo para simular los lanzamientos del dado, entre valores 1 y 6, esta comprende los siguientes parámetros:

- hight: Límite inferior del rango del que extraer el número aleatorio.
- low: Límite superior del rango del que extraer el número aleatorio.


```python
def Lanzamiento():
    return round(np.random.uniform(low = 0.5, high = 6.5, size = None),0)
    #como hace el man para redondear los lanzamientos 
```

### Función inicio juego
<br>
<div style="text-align: justify">Esta función permite inicializar el juego descontando 5.000 pesos a los jugadores y va cargando en el pozo lo que aporta cada uno en su respectiva mesa, el pozo está guardado en el DataFrame de nombre DP, las mesas se guardan en un diccionario llamado datos juego. Mientras que se recorre cada pozo de cada mesa le va descontando a cada jugador de cada mesa 5.000 pesos, si el jugador no tiene más de 5.000 no puede apostar y no puede jugar, mínimo debe haber mas de un jugador con más de 5.000 pesos para poder iniciar el juego. </div>

```python
def inicio_juego(DP,datos_juego):
    for y in range(len(DP)):
        recoleccion = 0
        ronda = datos_juego['Mesa-' + str(y + 1)]
        if sum(np.where(ronda["billetera jugador"] > 5000, 1, 0)) > 1:
            for row, data in ronda.iterrows():
                if(ronda.loc[row, "billetera jugador"] >= 5000):
                    ronda.loc[row, "billetera jugador"] = data["billetera jugador"] - 5000
                    recoleccion = recoleccion + 5000
                else:
                    print('el jugador'+ str(row) + 'del la mesa'+ str(y + 1) +'no tiene fondos para jugar')
                    #en este else entran los jugadores que ya no les alcanza para poner para el juego

        DP.loc[y, "pozo"] = recoleccion
```

#### Función de probabilidad 

Esta función simula 1000 lanzamientos, partiendo de estos hace un conteo de la frecuencia de cada uno de los resultados obtenidos en el dado y con base en esto calcula la probabilidad de sacar un numero mayor a cada numero que puede obtener el jugador en el dado, se necesita el parámetro del resultado del primer lanzamiento para que se pueda poner en marcha la función.


```python
def probabilidad(primer_lanzamiento):
    dice = [1,2,3,4,5,6]
    one = 0
    two=0
    three = 0
    four= 0
    five = 0
    six = 0
    for i in range(1000):# numero de simulacion
        result = random.choice(dice)
        if result == 1:
            one +=1
        elif result == 2:
            two +=1
        elif result == 3:
            three +=1
        elif result == 4:
            four +=1
        elif result == 5:
            five +=1
        elif result == 6:
            six +=1
    prob_trial_r=[one/1000,two/1000,three/1000,four/1000,five/1000,six/1000]
        #agregar el calculo de la probabilidad de los valores mayores al sacado
    if primer_lanzamiento== 2.0:
        prob=prob_trial_r[2]+prob_trial_r[3]+prob_trial_r[4]+prob_trial_r[5]

    if primer_lanzamiento== 3.0:
        prob=prob_trial_r[3]+prob_trial_r[4]+prob_trial_r[5]
        prob
    if primer_lanzamiento== 4.0:
        prob=prob_trial_r[4]+prob_trial_r[5]
        prob
    if primer_lanzamiento== 5.0:
        prob=prob_trial_r[5]
        prob
    return prob
```
